**According to usual practice we remove the column containing null values greater than 50% so we have to remove the column Unnamed Conatining 100% null values. For that we use drop() function.**

```python
Data=Data.drop("Unnamed: 11",axis=1)
```

**To Fill The Object type Null values we fill them with the mode of the values in that coloumn, to find mode we use mode() function.**

```python
State_Mode=Data['state'].mode()
State_Mode

```

**To fill the null values we use fillna() function.**

```python
Data["state"]=Data["state"].fillna("ca")
```

**Repeating same steps to clean our Dataset.**

```python
country_Mode=Data['country'].mode()
country_Mode
```

```python
Data['country']=Data['country'].fillna("us")
```

```python
City_Mode=Data['city'].mode()
City_Mode
```

```python
Data["city"]=Data["city"].fillna("seattle")
```

```python
Shape_Mode=Data['shape'].mode()
Shape_Mode
```

```python
Data["shape"]=Data["shape"].fillna("light")
```

```python
Shape_Mode=Data['shape'].mode()
Shape_Mode
```

**As Duration(seconds) and Duration(hours/min) containing negligible null values so we are removing the entire row containg null values. For That we use dropna() function.** 

```python
Data = Data.dropna(subset=['duration (seconds)', 'duration (hours/min)'])
```

```python
Comment_Mode=Data['comments'].mode()
Comment_Mode
```

**Now Using Matplolib and Seaborn Library to visualise our data.**










***Ploting the graph between the Number Of Shapes of UFOs Reported.***

```python
plt.figure(figsize=(20, 6))
sns.countplot(x=Data["shape"],palette='viridis')
plt.show()
```

***Plotting Line graph to observe shape of UFOs Observed in states.***

***Grouping The data of Country and shapes observed.***

```python
most_sightings = Data.groupby('country')['shape'].size().sort_values(ascending=False)
most_sightings
```

***PLotting the graph to visualise the above.***

```python
sns.barplot(most_sightings,palette='viridis')
plt.show()
```

```python
Data['duration (seconds)'] = pd.to_numeric(Data['duration (seconds)'], errors='coerce')
```

```python
sns.set(style="whitegrid", palette="pastel")

# Analyze the average encounter duration by country
avg_duration_by_country = Data.groupby('country')["duration (seconds)"].mean().sort_values()[:10]

# Plot the average encounter duration by country using Seaborn
plt.figure(figsize=(20, 6))
sns.barplot(x=avg_duration_by_country.values, y=avg_duration_by_country.index, palette="viridis")
plt.title('Average Encounter Duration by Country')
plt.xlabel('Average Encounter Duration (seconds)')
plt.ylabel('Country')
plt.show()
```

```python
shape_over_time = Data.pivot_table(index='state', columns='shape', aggfunc='size', fill_value=0)
```

```python
plt.figure(figsize=(25, 6))
for shape in shape_over_time.columns:
    sns.lineplot(x=shape_over_time.index, y=shape_over_time[shape], label=shape, linewidth=2.5)
plt.title('Distribution of UFO Shapes Over Time')
plt.xlabel('State')
plt.ylabel('Number of Sightings')
plt.show()
```
